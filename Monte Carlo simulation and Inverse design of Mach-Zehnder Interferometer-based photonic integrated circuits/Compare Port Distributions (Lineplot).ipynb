```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.rcParams['axes.labelsize'] = 14   # x/y axis labels
plt.rcParams['legend.fontsize'] = 12  # legend
plt.rcParams['xtick.labelsize'] = 13  # x ticks
plt.rcParams['ytick.labelsize'] = 13  # y ticks
```

# Insertion loss average

```python
dataset_paths = [ 
    "SweepData_0.1_0.05_1000_10x10.txt",
    "SweepData_0.3_0.05_1000_10x10.txt",
    "SweepData_0.5_0.05_1000_10x10.txt",
    "SweepData_0.7_0.05_1000_10x10.txt",
    "SweepData_0.9_0.05_1000_10x10.txt"
]
legend_labels = [
    "0.1 Insertion loss Range",
    "0.3 ",
    "0.5 ",
    "0.7 ",
    "0.9 ",
]



def db_to_watts(db_values):
    return 10 ** (db_values / 10)

def watts_to_db(watts_values):
    return 10 * np.log10(watts_values)

def load_and_process_data(file_path):
    # Load dataset assuming it is a CSV file
    data = pd.read_csv(file_path, sep='\s+', header=None)

    # Convert dB to watts for calculations
    data_watts = db_to_watts(data)

    # Calculate mean and standard deviation in watts
    mean_watts = data_watts.mean(axis=0)
    std_watts = data_watts.std(axis=0)

    # Convert mean and bounds back to dB
    mean_db = watts_to_db(mean_watts)
    upper_bound_db = watts_to_db(mean_watts + std_watts)
    lower_bound_db = watts_to_db(mean_watts - std_watts)

    return mean_db, upper_bound_db, lower_bound_db
```

```python
def plot_data_with_shading(dataset_paths, legend_labels,PlotName):
    plt.figure(figsize=(10, 6))

    for i, dataset_path in enumerate(dataset_paths):
        # Load and process each dataset
        mean_db, upper_bound_db, lower_bound_db = load_and_process_data(dataset_path)
        
        # Extract port indices for x-axis
        ports = np.arange(1, len(mean_db) + 1)

        # Plot the data with shaded variance
        plt.plot(ports, mean_db,linestyle="dashed",marker='o',markersize=10, label='-'+legend_labels[i])
        plt.fill_between(
            ports, 
            lower_bound_db, 
            upper_bound_db, 
            alpha=0.3
        )

    plt.xlabel("Ports")
    plt.ylabel("Power (dB)")
    #plt.yscale('log')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(PlotName,dpi=300, bbox_inches='tight')
    plt.show()

# Call the plotting function
plot_data_with_shading(dataset_paths, legend_labels, "Insertion loss range.png")

#Scatterplot
#enlarge
#Component random distribution
```

# Transmission Bar plot

```python
data = {
    "Target": [0.70108356, 1.71104704, 0.80656883, 0.85446338, 1.32830968, 1.22720252, 0.92619272, 0.35582394],
    "Result" : [.7096819433, 1.730764809, 0.8153131465, 0.860843303, 1.342087968, 1.240134414, 0.9373767918, 0.3615486626]
}
df = pd.DataFrame(data)

# Plot bar pairs for each row
x_labels = [f"Port {i+1}" for i in range(len(df))]
width = 0.4  # Width of each bar

# Create the bar plot
fig, ax = plt.subplots(figsize=(10, 7))

for i in range(len(df)):
    # Position the bars side by side
    ax.bar(i - width / 2, df.iloc[i, 0], width=width, label=f"Column1 - Row {i}", color="blue")
    ax.bar(i + width / 2, df.iloc[i, 1], width=width, label=f"Column2 - Row {i}", color="orange")

# Add labels and legend
ax.set_xticks(range(len(df)))
ax.set_xticklabels(x_labels)
ax.set_ylabel("Power (mW)")
ax.set_title("Target & Result transmission comparision")
plt.legend(["Target Transmission", "Result Transmission"], loc="upper right")
plt.savefig("Target & Result transmission comparision.png")
plt.show()

(df["Target"]-df["Result"])/df["Target"]
```
